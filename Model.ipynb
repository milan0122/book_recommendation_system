## Popularity Based Recommender System


```python
rating_book=rating.merge(book,on="ISBN")
rating_book
```

```python
num_rating_df=rating_book.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df=rating_book.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
#merge
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

I am taking only those books 250 highest rating books

```python
popular_df=popular_df[popular_df['num_ratings']>250].sort_values('avg_ratings',ascending=False)
popular_df.shape
```

```python
# let's again merge on book to get the publisher name, author name and image_url
popular_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-L','num_ratings','avg_ratings']]
```

## Collaborative Filtering Based Recommender System


```python
#let's find out first top reader and who give ratings in more than 200 books
X=rating_book.groupby('User-ID').count()['Book-Rating']>200
```

```python
#fetching the user-id 
top_readers=X[X].index
```

```python
filtered_rating=rating_book[rating_book['User-ID'].isin(top_readers)]
```

```python
filtered_rating
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_book=y[y].index
```

```python
famous_book
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_book)]
```

```python
final_ratings
```

```python
# writing in user id in column and all book in rows with rating values 
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
#filling with 0 in NAN values
pt.fillna(0,inplace=True)
pt
```

```python
#calculating_distance using cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(book_name):
    #fetch index
    index = np.where(pt.index==book_name)[0][0]
    """
    getting a list with index and distance 
    sorting the list in reverse order on basis of distance
    reterive only 5 highest similar books
    
    """
    similar_books=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    for i in similar_books:
        print(pt.index[i[0]])
   
```

```python
recommend('Message in a Bottle')
```
